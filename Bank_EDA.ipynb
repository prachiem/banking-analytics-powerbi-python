```python
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=localhost\\SQLEXPRESS;"
    "DATABASE=banking_case;"
    "Trusted_Connection=yes;"
)
```

```python
query = "SELECT * FROM customer"

df = pd.read_sql(query, conn)

df.head()
```

```python
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High']

df['Income Band'] = pd.cut(df['Estimated_Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
df.columns
```

```python
#Examine the Distribution of unique categories in categorical columns
categorical_cols = df[["BRId", "GenderId", "IAId", "Amount_of_Credit_Cards", "Nationality", "Occupation", "Fee_Structure", "Loyalty_Classification", "Properties_Owned", "Risk_Weighting", "Income Band"]].columns

for col in categorical_cols:
    print(f"Value Counts for '{col}':")
    display(df[col].value_counts())
```

#### Bivariate Analysis

```python
cols = ["BRId", "GenderId", "IAId", "Nationality", 
        "Occupation", "Fee_Structure", 
        "Loyalty_Classification", "Properties_Owned", 
        "Risk_Weighting", "Income Band"]

for col in cols:
    plt.figure(figsize=(6,4))
    
    sns.countplot(data=df, x=col, hue='Nationality')
    
    plt.title(f"{col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    
    plt.show()
    plt.close()
```

#### histogram of value counts for different Occupation

```python
for col in categorical_cols:
    if col =="Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title("Histogram of Occupation Count")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

####  Numerical Analysis

```python
numerical_cols = ['Estimated_Income', 'Superannuation_Savings', 'Credit_Card_Balance', 'Bank_Loans', 'Bank_Deposits', 'Checking_Accounts', 'Saving_Accounts', 'Foreign_Currency_Account'] 

#Univariate Analysis and visualization
plt.figure(figsize=(8,8))
for i, col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.tight_layout()   
plt.show()
```

## HeatMaps

```python
numerical_cols = ['Estimated_Income', 'Superannuation_Savings', 'Credit_Card_Balance', 'Bank_Loans', 'Bank_Deposits', 'Checking_Accounts', 'Saving_Accounts', 'Foreign_Currency_Account'] 

correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
